```python
df1=pd.DataFrame({'name':['John','Smith','Paul'],
                 'Age':['25','30','50']},
                index=[0,1,2])
df2=pd.DataFrame({'name':['Adam','Smith'],
                 'Age':['45','35']},
                index=[3,4])
```

```python
df3=pd.concat([df1,df2])
```

```python
df3
```

```python
df3.drop_duplicates('name')
```

```python
#Merging
```

```python
d1 = pd.DataFrame({'key':['K0','K1','K2','K3'],
                 'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3']
                  })
d2 = pd.DataFrame({'key':['K0','K1','K2','K3'],
                   'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
```

```python
pd.merge(d1,d2,how='inner',on='key')
```

```python
d1 = pd.DataFrame({'key':['K0','K1','K2','K3'],
                 'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3']
                  })
d2 = pd.DataFrame({'key':['K0','K1'],
                   'C':['C0','C1'],
                    'D':['D0','D1']})
```

```python
pd.merge(d1,d2,how='inner',on='key')
```

```python
pd.merge(d1,d2,how='outer',on='key')
```

```python
pd.merge(d1,d2,how='left',on='key')
```

```python
pd.merge(d1,d2,how='right',on='key')
```
